```python
import pandas as pd
from scipy.stats import ttest_rel

# Load the data from Excel file
df = pd.read_excel("C:/Users/ACER/Desktop/project_data.xlsx")

# Define the columns for summer, rainy, and winter measurements
summer_cols = ['Summer RR', 'Summer FEV/FVC', 'Summer PEFR', 'Summer MVV']
rainy_cols = ['Rainy RR', 'Rainy FEV/FVC', 'Rainy PEFR', 'Rainy MVV']
winter_cols = ['Winter RR', 'Winter FEV/FVC', 'Winter PEFR', 'Winter MVV']

# Calculate Mean and Standard Deviation (SD) for Continuous Variables
means_summer = df[summer_cols].mean()
std_devs_summer = df[summer_cols].std()
means_rainy = df[rainy_cols].mean()
std_devs_rainy = df[rainy_cols].std()
means_winter = df[winter_cols].mean()
std_devs_winter = df[winter_cols].std()

# Calculate Median and Interquartile Range (IQR) for Categorical Variables
medians_sex = df.groupby('Sex').median()
iqr_sex = df.groupby('Sex').quantile(0.75) - df.groupby('Sex').quantile(0.25)

# Perform paired t-tests for each continuous variable
results = {}
for i in range(len(summer_cols)):
    summer_col = summer_cols[i]
    rainy_col = rainy_cols[i]
    winter_col = winter_cols[i]
    
    # Perform paired t-test between summer and rainy
    summer_rainy_pval = ttest_rel(df[summer_col], df[rainy_col]).pvalue
    
    # Perform paired t-test between summer and winter
    summer_winter_pval = ttest_rel(df[summer_col], df[winter_col]).pvalue
    
    # Perform paired t-test between rainy and winter
    rainy_winter_pval = ttest_rel(df[rainy_col], df[winter_col]).pvalue
    
    # Store results
    results[f'{summer_col} vs. {rainy_col}'] = summer_rainy_pval
    results[f'{summer_col} vs. {winter_col}'] = summer_winter_pval
    results[f'{rainy_col} vs. {winter_col}'] = rainy_winter_pval

# Print mean and standard deviation for each season
print("Mean and Standard Deviation for Summer:")
print(means_summer)
print(std_devs_summer)
print("\nMean and Standard Deviation for Rainy:")
print(means_rainy)
print(std_devs_rainy)
print("\nMean and Standard Deviation for Winter:")
print(means_winter)
print(std_devs_winter)

# Print median and interquartile range for each sex
print("\nMedian and Interquartile Range for Sex:")
print(medians_sex)
print(iqr_sex)

# Print paired t-test results
print("\nPaired t-test Results:")
for comparison, pval in results.items():
    print(f'{comparison}: p-value = {pval}')
```
